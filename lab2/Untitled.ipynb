```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
center1 = np.array([2, 3])
center2 = np.array([8, 7])

n_points = 50
std_dev = 1.0

group1 = np.random.normal(loc=center1, scale=std_dev, size=(n_points, 2))
group2 = np.random.normal(loc=center2, scale=std_dev, size=(n_points, 2))
```

```python
plt.scatter(group1[:,0], group1[:,1])
plt.scatter(group2[:,0], group2[:,1])
```

```python
border = np.array([-1.0, 9.0])

X = np.linspace(0,10, 100)
Y = border[0]*X + border[1]

plt.plot(X,Y)
plt.scatter(group1[:,0], group1[:,1])
plt.scatter(group2[:,0], group2[:,1])
```

```python
points = np.concatenate((group1, group2))
targets = np.array([1.0]*50+[0.0]*50)
```

```python
border[0]*points[:,0] + border[1] - points[:,1]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
x = np.linspace(-10,10, 200)
y = sigmoid(x)

plt.plot(x,y)
```

```python
theta = border[0]*points[:,0] + border[1] - points[:,1]
output = sigmoid(theta)
```

```python
output
```

```python
loss = (output - targets)**2
```

```python
loss
```

```python
np.sum(loss)
```

```python
def train_step(points, targets, border, lr=0.1):
    output = sigmoid(border[0]*points[:,0] + border[1] - points[:,1])
    loss = (output - targets)**2
    loss = np.sum(loss)

    delta = np.random.normal(size=(2))
    border += lr*delta*loss

    return border
    
```

```python
border = np.array([-1.0, 8.0])

for i in range(1000):
    border = train_step(points, targets, border, 0.1)
```

```python
X = np.linspace(0,10, 100)
Y = border[0]*X + border[1]

plt.plot(X,Y)
plt.scatter(group1[:,0], group1[:,1])
plt.scatter(group2[:,0], group2[:,1])
```
